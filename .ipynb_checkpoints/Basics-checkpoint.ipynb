```python
import pandas as pd
df = pd.read_csv('executions.csv')
df.Race = pd.Categorical(df.Race)
df.County= pd.Categorical(df.County)
df
```

```python
df['rc'] = df.Race.cat.codes
df['cc'] = df.County.cat.codes
df
```

```python
df2 = df[["Race", "County", "Execution"]].copy()
df2 = df2.groupby(['County', 'Race']).count()
df2 = df2.reset_index()
df2
```

```python
df2["Execution"] = df2["Execution"].fillna(0.0)
```
